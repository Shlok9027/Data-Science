```python
import pandas as pd
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, root_mean_squared_error, r2_score

import numpy as np

model = LinearRegression()

data = pd.read_csv(r'C:\Users\shrey\SHLOK Python Code\MACHINE_LEARNING\student_sheet_1.csv')

X = data[['Hours']] # 2d input

y = data['Score'] # target column

model.fit(X,y)

predicted_score = model.predict(X)

mae = mean_squared_error(y,predicted_score)
mse = mean_squared_error(y, predicted_score)
rmse = np.sqrt(mse)

print("Mean Absolute Error : ", mae)
print("Mean Squared Error : ", mse)
print("Root Mean Square Error : ", rmse)


new_hour = float(input("Enter a Hours : "))

new_predict = model.predict([[new_hour]])

print(f'Predicted Score for {new_hour} Hours is : ', new_predict)
```
