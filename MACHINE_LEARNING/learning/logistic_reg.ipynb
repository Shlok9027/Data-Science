```python
from sklearn.linear_model import LogisticRegression

X = [[0],[1],[3],[4],[5],[6]]

y = [0,0,0,1,1,1]

model = LogisticRegression()
model.fit(X,y)



hours = float(input('Enter the hours you studied : '))

prediction_result = model.predict([[hours]])[0]


if prediction_result == 0:
    print('You will fail the exam.')
else:
    print('You will pass the exam.')

print(f'The predicted result is : {prediction_result}')

```
