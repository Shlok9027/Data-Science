```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


from sklearn.linear_model import LinearRegression

data = pd.read_csv(r"C:\Users\shrey\SHLOK Python Code\MACHINE_LEARNING\student_performance.csv")

X = data[['weekly_self_study_hours']]

y = data["total_score"]
model = LinearRegression()
model.fit(X,y)

predicted_score = model.predict(X)

mae = mean_absolute_error(y,predicted_score)
mse = mean_squared_error(y, predicted_score)
rmse = np.sqrt(mse)

r2 = r2_score(y, predicted_score)

print(f"Mean Absolute Error: {round(mae, 2)}")
print(f"Mean Squared Error: {round(mse, 2)}")
print(f"Root Mean Squared Error: {round(rmse, 2)}")
print(f"R-squared (model accuracy) :  {round(r2, 4)}")


plt.figure(figsize=(10,6))
plt.hist(data["total_score"], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Total Scores")
plt.xlabel("Total Score")
plt.ylabel("Numbet of Students")
plt.grid(True)
plt.grid(axis='y', alpha=0.75)
plt.show()

plt.figure(figsize=(10,6))
plt.scatter(X,y, color='Pink', label='Actual Scores')
plt.plot(X, predicted_score, color='blue', linewidth=2, label='Predicted Scores')
plt.title("model predictions vs Actual Scores")
plt.xlabel("Weekly Self Study Hours")
plt.ylabel("final Score")
plt.grid(True)
plt.grid(axis='y', alpha=0.75)
plt.show()

new_hour = [[5]]
predicted_new_score = model.predict(new_hour)
print(f'predicted_score for 5 hours of self study per week is: {round(predicted_new_score[0],2)}')
```
