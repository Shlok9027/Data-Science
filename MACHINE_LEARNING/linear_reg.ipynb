```python
from sklearn.linear_model import LinearRegression

X = [[1], [3],[5], [8], [10]]

y = [100,130,150,180, 220]


model = LinearRegression()

model.fit(X,y)

hours = float(input('Enter the Hours You Studied : '))

predict_mark = model.predict([[hours]])

print(f' as per the model you will score  {predict_mark} marks if you study for {hours} hours')
```
