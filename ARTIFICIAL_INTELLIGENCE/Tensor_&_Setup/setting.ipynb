```python
import torch 
x = torch.tensor([1., 2., 3., 4.])


y = torch.tensor([2., 4., 6., 8.])

w = torch.tensor(0.0)
b = torch.tensor(0.0)


y_pred = w * x + b

loss = ((y_pred - y) ** 2).mean()

print('Pridiction Values', y_pred)

print('Actual Value', y)

print("Loss", loss)
```
